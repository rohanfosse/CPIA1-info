```python
import numpy as np
import matplotlib.pyplot as plt


# DÃ©finition des valeurs de x avec un pas de 2
x_values = np.arange(-10, 11, 2)

y_values = []

for x in x_values:
    y_values.append(2*x**3 + 5*x - 10)

plt.figure(figsize=(8, 6))

plt.plot(x_values, y_values)
plt.xlabel('x')
plt.ylabel('y(x)')
plt.title('y(x) = 2x^3 + 5x - 10')
plt.show()
```

```python
import random

def simuler(n):
    l  = []
    for i in range(0,n):
        l.append(random.randint(1,6))

    return l

x = simuler(10)
print(x)

y = []

for i in range(1,max(x)+1):
    y.append(x.count(i))

plt.bar(range(1,len(y)+1),y,color='green')
plt.xlabel('Valeurs')
plt.ylabel('Occurences')
plt.title('Histogramme')
plt.show()
```

```python
def generer2(n,k):
    l = []
    for i in range(0,n+1):
        l.append(random.randint(k,6*k )) 

    return l


```

```python
def dictionnaire_carre(n):
    d = {}
    for i in range(1,n+1):
        d[i] = i**2
    return d

print(dictionnaire_carre(10))
```

```python
def dictionnaire_k(n,k):
    d = {}
    for i in range(1,n+1):
        d[i] = i**k
    return d
for k in range(0,11,2):
    print(dictionnaire_k(10,k))
    print()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# DÃ©finition de la fonction
def y(x):
    return 2*x**3 + 5*x - 10

# DÃ©finition des valeurs de x avec un pas de 2
x_values = np.arange(-10, 11, 2)
y_values = y(x_values)

# CrÃ©ation du graphique
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, 'ko-', label=r'$y(x) = 2x^3 + 5x - 10$')  # 'ko-' : points noirs avec ligne
plt.xlabel('x')
plt.ylabel('y(x)')
plt.title('ReprÃ©sentation graphique de la fonction')
plt.grid(True)
plt.legend()
plt.show()
```
