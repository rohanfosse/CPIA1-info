```python
import numpy as np
import matplotlib.pyplot as plt

x_values = np.arange(-10,11,2)

y_values = []

for x in x_values:
    y_values.append(2*x**3 + 5*x - 10)

plt.figure(figsize=(8,6))

plt.plot(x_values,y_values)
plt.xlabel('x')
plt.ylabel('y(x)')
plt.grid()
plt.title('y(x) = 2x^3 + 5x - 10')
plt.show()
```

```python
x_values = np.linspace(-np.pi,np.pi,100)

y_values = []

for x in x_values:
    y_values.append(2*np.sin(x) + 3*np.sin(2*x) + 2)

plt.plot(x_values,y_values,"r*")
plt.show()
```
