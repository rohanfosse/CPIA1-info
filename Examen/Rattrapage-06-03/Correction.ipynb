### **Étape 1 – Génération des températures**  
L’objectif est de créer une liste contenant **24 valeurs aléatoires** entre **-5°C et 30°C**, en utilisant **uniquement `random.uniform()`**.

```plaintext
DÉBUT

Définir n ← 24   // Nombre d'heures

Créer une liste temperatures vide

POUR i allant de 0 à n-1 FAIRE
    Générer une température aléatoire entre -5 et 30 avec random.uniform()
    Ajouter cette température à la liste temperatures
FIN POUR

AFFICHER "Températures générées :", temperatures

```
**Explication** :  
- On initialise une liste vide `temperatures`.  
- On remplit cette liste avec **24 valeurs aléatoires** comprises entre -5°C et 30°C.  
- L'affichage permet de vérifier les valeurs générées.  

---

### **Étape 2 – Calcul des valeurs minimales, maximales et moyennes**  
L’objectif est de **trouver la température minimale, maximale et la moyenne**, sans utiliser `min()`, `max()` ou `sum()`.

```plaintext
// Initialisation des variables
temp_min ← temperatures[0]
temp_max ← temperatures[0]
temp_sum ← 0

POUR CHAQUE température temp DANS temperatures FAIRE
    SI temp < temp_min ALORS
        temp_min ← temp   // Mise à jour de la température minimale
    FIN SI
    
    SI temp > temp_max ALORS
        temp_max ← temp   // Mise à jour de la température maximale
    FIN SI

    temp_sum ← temp_sum + temp  // Ajout de la température à la somme totale
FIN POUR

temp_moyenne ← temp_sum / n   // Calcul de la moyenne des températures

AFFICHER "Température minimale :", temp_min
AFFICHER "Température maximale :", temp_max
AFFICHER "Température moyenne :", temp_moyenne
```

**Explication** :  
- On initialise **temp_min** et **temp_max** avec la première valeur de la liste.  
- On **parcourt la liste** et met à jour ces valeurs lorsqu'on trouve un plus petit ou plus grand élément.  
- La **somme totale** est calculée manuellement.  
- La **moyenne** est obtenue en divisant cette somme par le nombre d’heures.  

---

### **Étape 3 – Détermination des heures des températures extrêmes**  
L’objectif est de **trouver l’heure où se situent la température minimale et maximale**, sans utiliser `index()`.

```plaintext
temp_min_heure ← 0
temp_max_heure ← 0

POUR i allant de 0 à n-1 FAIRE
    SI temperatures[i] = temp_min ALORS
        temp_min_heure ← i
        SORTIR DE LA BOUCLE   // On s'arrête dès qu'on trouve la première occurrence
    FIN SI
FIN POUR

POUR i allant de 0 à n-1 FAIRE
    SI temperatures[i] = temp_max ALORS
        temp_max_heure ← i
        SORTIR DE LA BOUCLE   // On s'arrête dès qu'on trouve la première occurrence
    FIN SI
FIN POUR

AFFICHER "Température minimale enregistrée à :", temp_min_heure, "h"
AFFICHER "Température maximale enregistrée à :", temp_max_heure, "h"
```

**Explication** :  
- On **parcourt la liste** et dès qu’on trouve une correspondance avec `temp_min`, on **enregistre l’indice** et on **sort de la boucle**.  
- On fait de même pour `temp_max`.  
- Cela permet d'éviter l'utilisation de `index()`.  

---

### **Étape 4 – Calcul de l’écart absolu moyen**  
L’objectif est d’appliquer la **formule donnée** pour calculer l’écart absolu moyen.

\[
\Delta T_{\text{moyen}} = \frac{|T_{\max} - T_{\min}|}{n-1}
\]

```plaintext
delta_T_moyen ← ABS(temp_max - temp_min) / (n - 1)

AFFICHER "Écart absolu moyen des températures :", delta_T_moyen
```

**Explication** :  
- On applique directement la **formule mathématique** donnée.  
- On utilise `ABS()` pour prendre la **valeur absolue**.  
- On divise par `(n - 1)`, car on considère **les variations entre chaque heure**.  

---

### **Étape 5 – Affichage des températures par heure**  
L’objectif est d’afficher les températures dans un **tableau clair**.

```plaintext
AFFICHER "Heure | Température (°C)"
AFFICHER "-------------------------"

POUR heure allant de 0 à n-1 FAIRE
    AFFICHER heure, "h  |", temperatures[heure]
FIN POUR
```

**Explication** :  
- On **affiche un tableau** où chaque ligne correspond à une heure de la journée.  
- L'affichage formaté permet une lecture plus simple.

````python
# Etape 0 : Importation des librairies

import random
import matplotlib.pyplot as plt
````

````python
# ## Étape 1 : Génération des températures
# Cette cellule génère 24 valeurs de températures aléatoires entre -5 et 30°C.

n = 24  # Nombre d'heures
temperatures = []

for i in range(n):
    temp = random.randint(-5, 30)  # Générer une température entière entre -5°C et 30°C
    temperatures.append(temp)

print("Températures générées :", temperatures)
````

````python
# ## Étape 2 : Calcul des valeurs minimales, maximales et moyennes

# On ne doit pas utiliser min(), max() ou sum().
temp_min = temperatures[0] # On initialise la température minimale à la première valeur de la liste
temp_max = temperatures[0] # On initialise la température maximale à la première valeur de la liste
temp_sum = 0

for temp in temperatures:
    if temp < temp_min:
        temp_min = temp
    if temp > temp_max:
        temp_max = temp
    temp_sum += temp

temp_moyenne = temp_sum / n

# Détermination des heures correspondantes
temp_min_heure = 0
temp_max_heure = 0
for i in range(n):
    if temperatures[i] == temp_min:
        temp_min_heure = i
        break
for i in range(n):
    if temperatures[i] == temp_max:
        temp_max_heure = i
        break
````

````python
# ## Étape 3 : Calcul de l'écart absolu moyen
delta_T_moyen = abs(temp_max - temp_min) / (n - 1)
````

````python
# ## Étape 4 : Affichage des résultats

print("Température minimale :", temp_min, "°C à", temp_min_heure, "h")
print("Température maximale :", temp_max, "°C à", temp_max_heure, "h")
print("Température moyenne sur la journée :", temp_moyenne, "°C")
print("Écart absolu moyen des températures :", delta_T_moyen, "°C")
````

````python
# ## Étape 5 : Affichage des températures par heure

print("\nHeure | Température (°C)")
print("-------------------------")
for heure in range(n):
    print(heure, "h  |", temperatures[heure])
````

````python
# ## Étape 6 : Tracé de la courbe des températures
plt.figure(figsize=(10, 5))  # Définition de la taille du graphique
plt.plot(range(n), temperatures, marker='o', linestyle='-', label='Température')  #
# - range(n) : Axe X représentant les heures (0 à 23)
# - temperatures : Axe Y représentant les températures
# - marker='o' : Ajoute des points sur chaque mesure
# - linestyle='-' : Trace une ligne entre les points
# - label='Température' : Légende pour identifier la courbe
plt.xlabel("Heure")  # Nom de l'axe X
plt.ylabel("Température (°C)")  # Nom de l'axe Y
plt.title("Évolution des températures sur 24h")  # Titre du graphique
plt.legend()  # Affichage de la légende
plt.grid()  # Ajout d'une grille pour une meilleure lisibilité
plt.show()  # Affichage du graphique de la courbe des températures
````

````python
# ## Étape 7 : Tracé de l'histogramme des températures
plt.figure(figsize=(8, 5))  # Définition de la taille de l'histogramme
plt.hist(temperatures, bins=8, edgecolor='black', alpha=0.7)  #
# - temperatures : Données utilisées pour l'histogramme
# - bins=8 : Nombre de classes pour regrouper les températures
# - edgecolor='black' : Ajout d'un contour noir aux barres
# - alpha=0.7 : Transparence des barres pour une meilleure visibilité
plt.xlabel("Température (°C)")  # Nom de l'axe X
plt.ylabel("Fréquence")  # Nom de l'axe Y
plt.title("Distribution des températures sur 24h")  # Titre de l'histogramme
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Ajout d'une grille uniquement sur l'axe Y
plt.show()  # Affichage de l'histogramme des températures
````
